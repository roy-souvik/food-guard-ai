```python
import sys
import json
import random
from pathlib import Path

# Ensure the parent directory is in the path to import your library
sys.path.insert(0, '..')
from foodguard_lib import MilkVisionPredictor, VAL_FOLDER, DB_PATH

print("--- 1. Initializing Predictor Pipeline ---")
predictor = MilkVisionPredictor()

# Automatically discover a valid sample image from your validation folders
val_dir = Path(VAL_FOLDER) / "val"
image_pool = list(val_dir.rglob("*.png")) + list(val_dir.rglob("*.jpg"))

if not image_pool:
    print(f"[Error] No validation images found in path: {val_dir}")
else:
    # Pick a random validation image file
    test_image = random.choice(image_pool)
    print(f"Selected Testing Asset: {test_image.name}")
    print(f"Absolute Path: {test_image.resolve()}\n")
    
    print("--- 2. Executing Dual-Mode Inference (DB + JSON Write) ---")
    # Execute with both flags switched to True
    payload = predictor.predict(
        image_path=str(test_image), 
        save_db=True, 
        save_json=True
    )
    
    print("\n--- 3. Returned Live Prediction Payload Object ---")
    print(json.dumps(payload, indent=4))
```
